```python
import pandas as pd

df = pd.read_excel("dataset.xlsx")

df.head()
```

```python
df["Item1"] = df["Item1"].str.lower().str.strip()
df["Item2"] = df["Item2"].str.lower().str.strip()

df = df.dropna()

print("Total rows:", len(df))
```

```python
interaction_lookup = {}

for _, row in df.iterrows():

    item1 = row["Item1"]
    item2 = row["Item2"]

    interaction_lookup[(item1, item2)] = {
        "type": row["Type"],
        "risk": row["Risk Level"],
        "interaction": row["Interaction"]
    }

print("Lookup created")
print("Total interactions:", len(interaction_lookup))
```

```python
import pickle

with open("interaction_lookup.pkl", "wb") as f:
    pickle.dump(interaction_lookup, f)

print("interaction_lookup.pkl saved")
```
